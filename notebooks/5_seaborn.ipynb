```python
import seaborn as sns
import matplotlib as plt
```

```python
# Use sns.get_dataset_names() see the available datasets in seaborn
# Let's use the restaurant tips data available in seaborn
tips = sns.load_dataset('tips')
tips.head()
```

# Categorical

## Count

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.countplot(x="day", data=tips, palette="Set2", hue="day")
plt.title("Count of Tips by Day of the Week")
plt.show()
# sns.countplot?
```

## Categorical Box plot

```python
import pandas as pd
tips["bin"] = pd.cut(tips["tip"], [0, 0.5, 1, 1.5, 2, 2.5, 3, 4, 5, 8])
tips
sns.boxplot(x="bin", y="tip", data=tips, palette="coolwarm", hue="time")
```

# BiVariate

Relacja pomidzy zmiennymi

```python
# Load the "iris" dataset
iris = sns.load_dataset("iris")

# Pair Plot
sns.set(style="ticks")
sns.pairplot(iris,diag_kind='hist', hue="species", markers=["o", "s", "D"])
```

# Heatmap

```python
correlation_matrix = iris.corr(numeric_only=True)

plt.figure(figsize=(10, 8))

# Create a heatmap of the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

```python
correlation_matrix = iris.corr(numeric_only=True)
plt.figure(figsize=(10, 8))

# Create a heatmap of the correlation matrix
sns.clustermap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```
